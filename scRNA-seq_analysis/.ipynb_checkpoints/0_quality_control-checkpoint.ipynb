## Quality control starts here

### Calculate mitochondrial content

Several tools automatically calculate the mitochondrial gene expression and many other features per cell. However, it is important to know how to calculate specially if you want to do more!

```r
# Identify the mitochondrial genes. Usually human and mouse starts with MT or Mt.
mt.genes <- rownames(alldata)[grep("^MT-", rownames(alldata))]

# Count matrix
C <- GetAssayData(object = alldata, slot = 'counts')

# Calculate the percentage
percent.mito <- colSums(C[mt.genes,]) / Matrix::colSums(C) * 100

# Add back to metadata information
alldata <- AddMetaData(alldata, percent.mito, col.name = "percent.mito")
```

## Calculate ribosomal content

Depending on your samples, you may want to verify the ribosomal proportion as well.

```r
# Identify the ribosomal genes. Usually starts with RP in human.
rb.genes <- rownames(alldata)[grep("^RP[SL]",rownames(alldata))]
length(rb.genes)

# Calculate the percentage
percent.ribo <- colSums(C[rb.genes,]) / Matrix::colSums(C) * 100

# Store back
alldata <- AddMetaData(alldata, percent.ribo, col.name = "percent.ribo")
```

## Decide where to cut

After a close look at your data, it is often recommended to remove some outliers cells. 

For example, in single cell experiments we can find cells with a high proportion of mitochondrial reads. IF, after removing those cells, still a good number of remaining cells that is ok. Another option is to either remove all mitochondrial reads from the dataset and see if the remaining gnees still have enough biological signal. And another option is to regress out the percent.mito variable during calling.

```r
# select cells with percent.mito less than a value
data.filt <- subset(alldata, subset = nFeature_RNA > 200 & nFeature_RNA < 2500 & percent.mito < 5)
```

```r
# plot violins for new data
VlnPlot(data.filt, features = "percent.mito")
```

## Look at technical bias with a different approach

https://bioconductor.org/packages/devel/bioc/vignettes/scater/inst/doc/overview.html
http://bioconductor.org/books/release/OSCA/quality-control.html#choice-of-qc-metrics

```r
sce <- as.SingleCellExperiment(alldata)
```

```r
# calculate all qc-metrics
df <- perCellQCMetrics(sce, subsets=list(Mito=mt.genes))
df
```

### option 1- Identifying low quality cells with fixed threshold

```r
#Remove cells with less than 1000 reads per library
qc.lib <- df$sum < 1e3  #1000 reads library size

#Express fewer than 300 genes
qc.nexprs <- df$detected < 3e2  #300 genes minimum

#Have mitochondrial proportion above 5%
qc.mito <- df$subsets_Mito_percent > 5

# total of cells to be removed
discard <- qc.lib | qc.nexprs | qc.mito
```

```r
# Summarize the number of cells removed for each reason.
DataFrame(LibSize=sum(qc.lib), NExprs=sum(qc.nexprs), MitoProp=sum(qc.mito), Total=sum(discard))
```

###  option 2- With adaptive thresholds

To obtain an adaptive threshold, we assume that most of the dataset consists of high-quality cells.
We then identify cells that are outliers for the various QC metrics, based on the median absolute deviation (MAD) from the median value of each metric across all cells. Specifically, a value is considered an outlier if it is more than 3 MADs from the median in the “problematic” direction. This is loosely motivated by the fact that such a filter will retain 99% of non-outlier values that follow a normal distribution.

```r
qc.lib2 <- isOutlier(df$sum, log=TRUE, type="lower")
qc.nexprs2 <- isOutlier(df$detected, log=TRUE, type="lower")
attr(qc.lib2, "thresholds")
```
